```python
danceability = tracks_df ['danceability']
energy = tracks_df['energy']

bins = 10 

fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot( projection = '3d')

hist, xedges, yedges = np.histogram2d(danceability, energy, bins = bins)

xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing = 'ij')
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort = 'average', color = 'maroon', edgecolor = 'black')
ax.set_xlabel('Danceability')
ax.set_ylabel('Energy')
ax.set_zlabel('Count')
ax.set_title("3D Histogram of Danceability and Energy")

plt.show()
```

```python
from matplotlib import cm
x = tracks_df['danceability']
y = tracks_df['energy']
z = tracks_df['loudness']

x_grid, y_grid = np.meshgrid(np.linspace(x.min(),x.max(),100),
                             np.linspace(y.min(), y.max(),100))

z_grid = np.interp(np.sqrt(x_grid ** 2 + y_grid ** 2),
                   np.sqrt(x ** 2 + y **2),
                   z)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(x_grid, y_grid, z_grid)

ax.set_box_aspect([1,1,1])

ax.set_xlabel('Danceability')
ax.set_ylabel('Energy')
ax.set_zlabel('Loudness')
ax.set_title('3D Surface Plot of Danceability, Energy, and Loudness')

plt.show()
```
